```python
#Relationship between Monthly Charges and Total Charges
sns.lmplot(data=df1_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

Total Charges increase as Monthly Charges increase - as expected.

```python
# Define the columns you want to create KDE plots for
columns = ['MonthlyCharges', 'TotalCharges']

# Create a KDE plot for each column
for col in columns:
    plt.figure(figsize=(10, 6))
    for churn_status in [0, 1]:
        sns.kdeplot(df1_dummies[col][df1_dummies["Churn"] == churn_status],
                    shade=True, label="Churn" if churn_status == 1 else "No Churn",
                    color="Blue" if churn_status == 1 else "Red")

    plt.legend(loc='upper right')
    plt.ylabel('Density')
    plt.xlabel(col)
    plt.title(f'{col} by Churn')

    plt.show()
```

Churn is high when Monthly Charges ar high

higher Churn at lower Total Charges

However if we combine the insights of 3 parameters i.e. Tenure, Monthly Charges & Total Charges then the picture is bit clear :- Higher Monthly Charge at lower tenure results into lower Total Charge. Hence, all these 3 factors viz Higher Monthly Charge, Lower tenure and Lower Total Charge are linkd to High Churn.

We can clearly see our target dataset is highly imbalanced and if we use our any of classification models it's not give accurate results. This data imbalance can be dealt using resampling methods; upsampling and downsampling.

In upsampling or oversampling we increase the minority class to match with the majority class.
In downsampling or undersampling some data has been deleted from the majority class randomly to match it with the minority class data.
Here we will be using SMOTE (Synthetic Minority Oversampling Technique) to resample this data.

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd
# Create an empty DataFrame to store the metrics
model_metrics = pd.DataFrame(columns=["Model", "Data Type", "Accuracy", "Precision", "Recall", "F1-Score", "ROC AUC"])

# Define the list of models for both datasets
models = [
    ("Logistic Regression", lgmodel, X_test, "Original Data"),
    ("Decision Tree", dtmodel, X_test, "Original Data"),
    ("Random Forest", rfmodel, X_test, "Original Data"),
    ("Ada Boost", adamodel, X_test, "Original Data"),
    ("Logistic Regression with SMOTE", lgmodel_smote, Xr_test, "SMOTE Data"),
    ("Decision Tree with SMOTE", dtmodel_smote, Xr_test, "SMOTE Data"),
    ("Random Forest with SMOTE", rfmodel_smote, Xr_test, "SMOTE Data"),
    ("Ada Boost with SMOTE", adamodel_smote, Xr_test, "SMOTE Data")
]

# Calculate and store the evaluation metrics for each model and data type
for model_name, model, X, data_type in models:
    y_pred = model.predict(X)
    accuracy = accuracy_score(y_test if data_type == "Original Data" else yr_test, y_pred)
    precision = precision_score(y_test if data_type == "Original Data" else yr_test, y_pred)
    recall = recall_score(y_test if data_type == "Original Data" else yr_test, y_pred)
    f1 = f1_score(y_test if data_type == "Original Data" else yr_test, y_pred)
    roc_auc = roc_auc_score(y_test if data_type == "Original Data" else yr_test, model.predict_proba(X)[:, 1])
    
    model_metrics = model_metrics.append({
        "Model": model_name,
        "Data Type": data_type,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1,
        "ROC AUC": roc_auc
    }, ignore_index=True)

# Sort the DataFrame by a chosen metric, e.g., accuracy
model_metrics = model_metrics.sort_values(by="Accuracy", ascending=False)

# Reset the index for better visualization
model_metrics.reset_index(drop=True, inplace=True)

# Display the DataFrame with evaluation metrics
model_metrics
```

When considering the SMOTE-enhanced dataset, models such as Ada Boost with SMOTE and Logistic Regression with SMOTE perform well in terms of accuracy, precision, recall, F1-score, and ROC AUC. They are suitable for handling class imbalance and achieving good predictive performance.
On the original data, the models struggle with lower recall, which indicates a difficulty in capturing the minority class. SMOTE is effective in addressing this issue and improving model performance.
The choice of model should depend on your specific goals and the importance of precision, recall, and other metrics in your application. In scenarios where recall is crucial (e.g., minimizing false negatives), the SMOTE-enhanced models may be more appropriate.
